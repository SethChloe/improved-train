```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
from pandas import DataFrame,Series
```

```python
starbucks=pd.read_csv('starbucks.csv')
starbucks.head()
```

```python
starbucks.isnull().sum()
```

```python
starbucks['Brand'].unique()
```

```python
ret=starbucks['Country'].unique()
ret
```

```python
len(ret)
```

```python
country_count=starbucks['Country'].value_counts()[0:10]
country_count
```

```python
plt.style.use('ggplot')
labels=list(country_count.index)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Country Top 10')
plt.bar(range(len(labels)),country_count)
plt.xticks(range(len(labels)),labels,fontsize=12)
```

```python
city_count=starbucks['City'].value_counts()[:10]
city_count
```

```python
plt.figure(figsize=(10,6))
plt.rcParams['font.sans-serif']=['simhei']
# plt.rcParams['axes.unicode_minus']=False
labels=list(city_count.index)
plt.ylabel('Count')
plt.xlabel('City')
plt.title('City Top 10')
plt.bar(range(len(labels)),city_count)
plt.xticks(range(len(labels)),labels)
```
